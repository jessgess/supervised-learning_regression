```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import time
from datetime import timedelta, datetime, date
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import Ridge, Lasso, LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import GridSearchCV, train_test_split, KFold
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler, MinMaxScaler
pd.set_option('display.max_columns', 400) 
```

```python
data = pd.read_csv('data/nfl.csv')
print(data.shape)
data.head()
```

```python
for col in list(data):
    print('{} has {} NaNs'.format(col, data[col].isnull().sum())) 
```

```python
for col in list(data):
    print('{} has {} unique values--> {}'.format(col, data[col].nunique(), list(data[col].unique())))
```

```python
data[(data['Cntry'] == 'ITA') | (data['Cntry'] == 'EST')]['Nat']
```

```python
#creating age column for each pleayer 
to_date_time = lambda x: pd.to_datetime(x, format='%y-%m-%d')
data['Born'] = data['Born'].apply(to_date_time)
data['Age'] = (date.today() - data['Born'])/365.25
data['Age'] = pd.to_numeric(data['Age'].dt.days)
```

```python
le = LabelEncoder()
data.Hand = le.fit_transform(data.Hand)
```

```python
data = pd.get_dummies(data, columns=['Nat', 'Position', 'Team'])
```

```python
print(data.shape)
#data.dropna(axis=0, inplace=True)
data.drop(['Unnamed: 0', 'Born', 'City', 'Pr/St', 'Cntry', 'Last Name', 'First Name'], axis=1, inplace=True)
data.shape
```

```python
draft = ['DftYr', 'DftRd', 'Ovrl']
for col in draft:
    data[col].replace(np.nan, 0.0, inplace=True)
```

```python
data.dropna(axis=0, inplace=True)
data.shape
```

```python
featurematrix = data.copy()
```

#### There are two very high salaries (outlier/leverage points) as below, removing those increses the R_squared by 0.02 

```python
print(data.Salary.max())
data[data.Salary == 13800000].index
```

```python
data2 = data[data['Salary'] < 13800000]
data2.shape
```

```python
sns.boxplot(y=data2.Salary)
```

```python
featurematrix = data2.copy()
```

#### Developing models:

```python
#results are very unstable, sometimes standard scaler gives better prediction sometimes minmax, so trying with both.
scaler = MinMaxScaler()
featurematrix_mm = pd.DataFrame(
                            scaler.fit_transform(featurematrix), 
                            columns = featurematrix.columns, 
                            index = featurematrix.index)
```

```python
scaler = StandardScaler()
featurematrix_sc = pd.DataFrame(
                            scaler.fit_transform(featurematrix), 
                            columns = featurematrix.columns, 
                            index = featurematrix.index)
```

```python
X = featurematrix_mm.iloc[:, 1:]
y = featurematrix_mm['Salary']
```

```python
linear_reg = LinearRegression(normalize=True)
linear_reg.fit(X, y)
y_pred = linear_reg.predict(X)
res = y - y_pred
```

```python
centered_response = (y - y.mean())

RSS = np.linalg.norm(res)**2
TSS = np.linalg.norm(centered_response)**2

print("R-squared: {0:.4f}".format(1-RSS/TSS)) #good results should be close to 1 so below is ok
```

```python
#checking prediction vs actual values
print("predicted salary:", np.dot(X.loc[121], linear_reg.coef_) + linear_reg.intercept_)
print("actual salary (scaled):", y.loc[121])
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(y, res)
plt.xlabel("Response (Salary)", fontsize=12)
plt.ylabel("Residuals", fontsize=12)
```

```python
linear_reg_coefs = linear_reg.coef_

plt.figure(figsize=(30, 10))

plt.stem(np.arange(0,len(linear_reg_coefs)), linear_reg_coefs)
plt.xlabel("Index of coefficient", fontsize=12)
plt.ylabel("Magnitude of coefficient", fontsize=12)
```

```python
#tried using only the important features (feat>7le7 feat<-7le7 from above graph) as training set:
#X_reduced = featurematrix[important_feats] 
#as it turns out you cant just ditch other features witout changing the magnitude of the reamianing coefs, so got worst results. 
coefs = linear_reg_coefs.tolist()
all_feats = list(X)
coefs_feats = list(zip(coefs, all_feats))
sorted_coef = sorted(coefs_feats)
del sorted_coef[31:-10] #the first 30 and the last 9 have the largest coefs so only keeping those
important_feats = [x[1] for x in sorted_coef]
important_feats
```

```python
X = featurematrix.iloc[:, 1:] #all data with one-hot encodes 
y = featurematrix['Salary']
```

```python
ridge = Ridge(alpha=1, normalize=True)
ridge.fit(X, y)

normal_lr = Ridge(alpha=0, normalize=True)
normal_lr.fit(X, y)

feature_names = list(X)
width = 0.4
plt.figure(figsize=(30, 6))
xaxis = np.arange(len(feature_names))
plt.bar(xaxis-width/2, ridge.coef_, width=width, label="With penalty")
plt.bar(xaxis+width/2, normal_lr.coef_, width=width, label="Without penalty")
plt.xticks(xaxis, feature_names)
plt.legend(fontsize=12);
```

```python
import warnings
warnings.filterwarnings('ignore')

lasso_gcv = GridSearchCV(Lasso(), {'alpha': np.logspace(-4, 2, 10)}, cv=10)
lasso_gcv.fit(X, y)
ridge_gcv = GridSearchCV(Ridge(), {'alpha': np.logspace(-4, 5, 20)}, cv=10)
ridge_gcv.fit(X, y)

RSS1 = np.linalg.norm(y - lasso_gcv.predict(X))**2
RSS2 = np.linalg.norm(y - ridge_gcv.predict(X))**2
TSS  = np.linalg.norm(y - y.mean())**2

print("R2 of the best LASSO: {0:.3f}".format(1-RSS1/TSS))
print("R2 of the best RIDGE: {0:.3f}".format(1-RSS2/TSS))
```

```python
lasso_gcv.best_estimator_
```
